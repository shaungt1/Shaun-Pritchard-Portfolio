```python
# Imports for decoding and analysiing data
import pandas as pd
import numpy as np
import io
from datetime import date
from pyais.stream import FileReaderStream
```

```python
# Add file to be decoded here:
AISFile = "/content/AIS_Test_NMEA.txt"
```

```python
columns = [
'Msg_type',
'Repeat',
'MMSI',
'Nav_status',
'Turn',
'Speed',
'Accuracy',
'longitude',
'Latitude',
'Course',
'heading',
'second',
'Maneuver',
'spare',
'Raim',
'Radio',
]
```

```python
# Loop Through ASI file to decode and add all rows to a list
datalist = []
for msg in FileReaderStream(AISFile):
    decoded = msg.decode()
    datalist.append(decoded)
# Output list to dataframe
df = pd.DataFrame(datalist, columns=['data'])
print(df.head(10))
```

```python
# Output decoded data to csv File so each value is in a cell
df.to_csv('Temp/decoded_ASI_NH.csv', sep='|', index=False)
print('Decoded ASI File with no headers is complete in Temp Folder')
```

```python
InputFile = 'Temp/decoded_ASI_NH.csv'
df1 = df
new_header = df1.iloc[0] #grab the first row for the header
df1 = df1[1:] #take the data less the header row
df1.columns = new_header #set the header row as the df header
df1 = pd.read_csv(InputFile, names=columns)
df1 = df1.drop(df1.index[0])
df1.head(5)
```

```python
outPutFile = "AIS_DECODED_DATA"
date = date.today()
tm = date.strftime("%m_%d_%y")
df1.to_csv(outPutFile + '/decoded_ASI_' + tm +'.csv', index=False)
print('Final ASI file has been decoded and sent to csv file')
```
